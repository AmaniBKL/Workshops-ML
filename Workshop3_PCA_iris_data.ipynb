```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names = ['sepal length', 'sepal width', 'petal length', 'petal width', 'target'])
```

```python
df.head()
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
x = df.loc[:, features].values
```

```python
x
```

```python
y = df.loc[:, ['target']].values
y
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x_dataframe = pd.DataFrame(data=x, columns=features)
```

```python
x_dataframe.head()
```

```python
pca = PCA(n_components=2)
p = pca.fit_transform(x)
```

```python
p_dataframe = pd.DataFrame(data=p, columns=['1st Dim', '2nd Dim'])
```

```python
p_dataframe.head()
```

```python
final_df = pd.concat([p_dataframe, df[['target']]], axis=1)
```

```python
final_df.head()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.set_xlabel('1st Dim')
ax.set_ylabel('2nd Dim')
ax.set_title('PCA')

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']

for i, j in zip(targets, colors) : 
  ax.scatter(final_df.loc[final_df['target']==i, '1st Dim'], final_df.loc[final_df['target']==i, '2nd Dim'], c=j, s=50)

ax.legend(targets)
ax.grid()
```
